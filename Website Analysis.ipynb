# Website Analysis Performance Project

### Import csv file


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data-export.csv")
df.head()
```

```python
df.columns = df.iloc[0]
df = df.drop(index = 0).reset_index(drop = True)
df.columns = ["Primary channel", "Date_hour", "Users", "Sessions", "Engaged sessions", "Average engagement time per session", "Engaged sessions per use", "Events per session", "Engagement rate", "Event count"]
```

```python
df.head()
```

```python
df.info()
```

```python
df["Date_hour"] = pd.to_datetime(df["Date_hour"], format = "%Y%m%d%H", errors = "coerce")
df.head()
```

```python
numeric_cols = df.columns.drop(["Primary channel", "Date_hour"])
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors = 'coerce')
df["Hours"] = df["Date_hour"].dt.hour
```

```python
df.info()
```

```python
df.describe()
```

## Visualization Charts

### 1. Session and users over time

```python
df.groupby("Date_hour")[["Sessions", "Users"]].sum().plot(
    figsize=(20, 7)
)

plt.title("Session and Users Over Time")
plt.xlabel("Date_Hour")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

### 2. Total users by channel

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df, x="Primary channel", y="Users", estimator=sum, hue="Primary channel", palette="viridis", legend=False)
plt.title("Total Users by Channel")
plt.xticks(rotation=45)
plt.show()
```

### 3. Average Engagement time by Channel


```python
plt.figure(figsize=(8,4))
sns.barplot(data=df, x="Primary channel", y="Average engagement time per session", estimator=np.mean, hue="Primary channel", 
            palette="magma", legend=False)
plt.title("Average Engagement Time by Channel")
plt.xticks(rotation=45)
plt.show()
```

### 4. Engagement Rate Distribution by channel


```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Primary channel", y="Engagement rate", hue="Primary channel", palette="coolwarm")
plt.title("Engagement Rate Distribution by Channel")
plt.xticks(rotation=45)
plt.show()
```

### 5. Engaged vs non engaged sessions

```python
session_df = df.groupby(["Primary channel"])[["Sessions", "Engaged sessions",]].sum().reset_index()
session_df["Non-Engaged"] = session_df["Sessions"] - session_df["Engaged sessions"]
session_df_melted = session_df.melt(id_vars="Primary channel", value_vars=["Engaged sessions", "Non-Engaged"])

plt.figure(figsize=(8,5))
sns.barplot(data=session_df_melted, x="Primary channel", y="value", hue="variable")
plt.title("Engaged VS Non-Engaged Sessions")
plt.xticks(rotation=45)
plt.show()
```

### 6. Traffic by hour and channel

```python
heat_map = df.groupby(["Hours", "Primary channel"])["Sessions"].sum().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(heat_map, cmap="YlGnBu", linewidths=.5, annot=True, fmt=".0f")
plt.title("Traffic by Hour and Channel")
plt.xlabel("Primary channel")
plt.ylabel("Hour of Day")
plt.show()
```

### 7. Engagement rate vs session over time

```python
df_plot = df.groupby("Date_hour")[["Engagement rate", "Sessions"]].mean().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(df_plot["Date_hour"], df_plot["Engagement rate"], label="Engagement rate", color="green")
plt.plot(df_plot["Date_hour"], df_plot["Sessions"], label="Sessions", color="blue")
plt.title("Engagement Rate vs Sessions Rate")
plt.xlabel("Date Hour")
plt.show()
```
